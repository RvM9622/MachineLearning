```python
from imblearn.over_sampling import SMOTE

x = df.drop(columns=['Diagnosis'])
y = df['Diagnosis']

smote = SMOTE(sampling_strategy='minority')
x_resampled, y_resampled = smote.fit_resample(x, y)

df = pd.concat([x_resampled, y_resampled], axis = 1)

print(df['Diagnosis'].value_counts())
```

```python
x = df.drop(['Diagnosis'], axis = 1)
y = df['Diagnosis']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier

model_lr = LogisticRegression()
model_nb = GaussianNB()
model_knn = KNeighborsClassifier()
model_dt = DecisionTreeClassifier()
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

ml_names = ['Logistic Regression', 'Naive Bayes', 'KNN', 'Decision Tree']
ml_model = [model_lr, model_nb, model_knn, model_dt]

for names, models in zip(ml_names, ml_model):
  models.fit(x_train, y_train)
  y_pred = models.predict(x_test)
  print(f"{names} Accuracy Score : {accuracy_score(y_test, y_pred)*100:.2f} %")
```

```python
for names, models in zip(ml_names, ml_model):
  y_pred = models.predict(x_test)
  cm = confusion_matrix(y_test, y_pred)
  print(f'Confusion Matrix for {names}:\n{cm}')

  cr = classification_report(y_test, y_pred)
  print(f'Classification Report for {names}:\n{cr}')
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier

ada_model = AdaBoostClassifier(n_estimators=200, random_state = 42)
gr_model = GradientBoostingClassifier(n_estimators = 200, random_state = 42)
rf_model = RandomForestClassifier(n_estimators = 200, random_state = 42)

ada_model.fit(x_train, y_train)
gr_model.fit(x_train, y_train)
rf_model.fit(x_train, y_train)

y_pred_ada = ada_model.predict(x_test)
y_pred_gr = gr_model.predict(x_test)
y_pred_rf = rf_model.predict(x_test)
```

```python
ada_score = accuracy_score(y_test, y_pred_ada)
gr_score = accuracy_score(y_test, y_pred_gr)
rf_score =accuracy_score(y_test, y_pred_rf)
print(f'AdaBoost Accuracy: {ada_score*100:.2f} %')
print(f'Gradient Boosting Accuracy: {gr_score*100:.2f} %')
print(f'Random Forest Accuracy: {rf_score*100:.2f} %')
```

```python
cm = confusion_matrix(y_test, y_pred_rf)
print(f'Confusion Matrix for {names}:\n{cm}')

cr = classification_report(y_test, y_pred_rf)
print(f'Classification Report for {names}:\n{cr}')
```
