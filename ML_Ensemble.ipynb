```python
import pandas as pd

df = pd.read_csv("diabetes.csv")
```

```python
df.info()
```

```python
df
```

```python
x = df.drop("Outcome", axis=1)
y = df["Outcome"]
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier

bag_model = BaggingClassifier()
bag_model.fit(x_train, y_train)
bag_model.score(x_test, y_test)
```

```python
bag_model = BaggingClassifier(n_estimators=2000)
bag_model.fit(x_train, y_train)
bag_model.score(x_test, y_test)
```

```python
tree = DecisionTreeClassifier()
bag_model = BaggingClassifier(estimator = tree, n_estimators=2000)
bag_model.fit(x_train, y_train)
bag_model.score(x_test, y_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

y_pred = bag_model.predict(x_test)
cm = confusion_matrix(y_test, y_pred)
print(cm)
print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=2000)
model.fit(x_train, y_train)
model.score(x_test, y_test)
```

```python
y_pred = model.predict(x_test)
cm = confusion_matrix(y_test, y_pred)
print(cm)
print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import AdaBoostClassifier

model_ada = AdaBoostClassifier(n_estimators=2000)
model_ada.fit(x_train, y_train)
model_ada.score(x_test, y_test)
```

```python
y_pred = model_ada.predict(x_test)
cm = confusion_matrix(y_test, y_pred)
print(cm)
print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import GradientBoostingClassifier

model_gb = GradientBoostingClassifier(n_estimators=2000)
model_gb.fit(x_train, y_train)
model_gb.score(x_test, y_test)
```

```python
y_pred = model_gb.predict(x_test)
cm = confusion_matrix(y_test, y_pred)
print(cm)
print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier

estimators = []
model1 = LogisticRegression()
estimators.append(('logistic', model1))
model2 = DecisionTreeClassifier()
estimators.append(('cart', model2))
model3 = SVC()
estimators.append(('svm', model3))

model_voting = VotingClassifier(estimators, voting='hard')
model_voting.fit(x_train, y_train)
```

```python
model_voting.score(x_test, y_test)
```
