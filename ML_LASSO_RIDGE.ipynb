```python
import pandas as pd

df = pd.read_csv('Melbourne_housing_FULL.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.nunique()
```

```python
col = ['Suburb', 'Rooms', 'Type', 'Method', 'Regionname', 'Propertycount', 'Distance', 'CouncilArea', 'Bedroom2', 'Bathroom', 'Car',
        'Landsize', 'BuildingArea', 'Price' ]

df = df[col]
```

```python
df
```

```python
df.dropna(subset=['Price'], inplace=True)
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
cols = ['Bathroom','Bedroom2','Car','Distance', 'Propertycount']
df[cols] = df[cols].fillna(0)
```

```python
df.isna().sum()
```

```python
df['Landsize'] = df['Landsize'].fillna(df.Landsize.mean())
df['BuildingArea'] = df['BuildingArea'].fillna(df.BuildingArea.mean())
```

```python
df.isna().sum()
```

```python
df.dropna(inplace = True)
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
df = pd.get_dummies(df, drop_first= True)
```

```python
df.head()
```

```python
df.info()
```

```python
y  = df['Price']
x = df.drop('Price', axis = 1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 2)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
model.score(x_test, y_test)
```

```python
model.score(x_train, y_train)
```

```python
from sklearn.linear_model import Lasso

lasso_reg = Lasso(alpha = 100, max_iter=950)
lasso_reg.fit(x_train, y_train)
```

```python
lasso_reg.score(x_test,y_test)
```

```python
lasso_reg.score(x_train,y_train)
```

```python
from sklearn.linear_model import Ridge

ridge_reg = Ridge(alpha = 100, max_iter=950)
ridge_reg.fit(x_train, y_train)
```

```python
ridge_reg.score(x_test, y_test)
```

```python
ridge_reg.score(x_train, y_train)
```
